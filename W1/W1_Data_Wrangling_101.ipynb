## Import Libraries

```python
import numpy as np
import pandas as pd
```

## Load Datasets

```python
cust_csv_path = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQxcCDTAalMyu4Tc_eelskRhZXldH9TSliZLFcW39_LlzOkIReCRdK0V7JR6APjJH-EluveA5c-m237/pub?gid=0&single=true&output=csv"
churn_csv_path = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQxcCDTAalMyu4Tc_eelskRhZXldH9TSliZLFcW39_LlzOkIReCRdK0V7JR6APjJH-EluveA5c-m237/pub?gid=755674277&single=true&output=csv"
```

```python
cust_df = pd.read_csv(cust_csv_path)
```

```python
churn_df = pd.read_csv(churn_csv_path)
```

```python
cust_df
```

```python
churn_df
```

## Merge Datasets

```python
df = pd.merge(cust_df, churn_df, how="left", on="cust_id")
```

```python
df
```

## Drop Missing Values

```python
df.info()
```

```python
df.dropna(subset=["churn"], inplace=True)
```

```python
df
```

```python
df.info()
```

## Cap the Outliners

```python
df.boxplot(column="height")
```

```python
df.dropna().sort_values('height')[['height']].plot.bar(use_index=False, figsize=(20,3), xticks=[], xlim=(0,1200), ylim=(100,500))
```

```python
low, high = df["height"].quantile([0.05, 0.95])
print(low, high)
```

```python
df.loc[df[df["height"]<low].index,"height"]=low
```

```python
df.loc[df[df["height"]>high].index,"height"]=high
```

```python
df.boxplot(column="height")
```

**Quiz !!**

> ลอง cap ที่ percentile 10-90

```python
low, high = df["height"].quantile([0.1, 0.9])
print(low, high)
```

```python
df.loc[df[df["height"]<low].index,"height"]=low
df.loc[df[df["height"]>high].index,"height"]=high

df.boxplot(column="height")
```

## Extract Date Time

```python
df["birth"]
```

```python
df["birth"] = pd.to_datetime(df["birth"], format='%m/%d/%Y')
```

```python
df["birth"]
```

```python
df["birth"].dt.year
```

```python
2025 - df["birth"].dt.year
```

```python
df["age"] = 2025 - df["birth"].dt.year
```

```python
df
```

**Quiz !!**

> สร้างคอลัมน์ birth_dow เพื่อแสดงวันเกิดในสัปดาห์ sun, mon, tue, wed, thu, fri, sat

```python
dow_map = {0: "mon", 1: "tue", 2: "wed", 3: "thu", 4: "fri", 5: "sat", 6: "sun"}
df["birth_dow"] = df["birth"].dt.dayofweek.map(dow_map)
df
```

## Bining

```python
df["age"].hist()
```

```python
cut_labels = ['young', 'teen', 'adult', 'senior']
cut_bins = [0, 13, 25, 61, 100]
df['age_group'] = pd.cut(df['age'], bins=cut_bins, labels=cut_labels)
```

```python
df
```

```python
df.boxplot(column="age", by="age_group")
```

```python
df.boxplot(column="salary", by="age_group")
```

**Quiz !!**

> ทดลองทำ binning ของกลุ่มเงินเดือน Low (น้อยกว่า 50,000), Med (50,000 - 100,000), High (มากกว่า 100,000)

```python
# Answer
cut_labels = ['low', 'med', 'high']
cut_bins = [0, 50000, 100000, float("inf")]
df['salary_group'] = pd.cut(df['salary'], bins=cut_bins, labels=cut_labels)
df[['salary_group', 'salary']]
```
